# Add Data Column

```python
import pandas as pd
data = pd.read_csv('datas/beijing_tianqi/beijing_tianqi_2018.csv')

data.head()
```

```python
# Delete °C
data.loc[:, 'bWendu'] = data['bWendu'].str.replace('℃','').astype('int32')
data.loc[:, 'yWendu'] = data['yWendu'].str.replace('℃','').astype('int32')
```

Caculate temperature

```python
data.loc[:, 'wencha'] = data['bWendu'] - data['yWendu']
```

```python
data.head()
```

## .apply

Apply a function along an axis of the Dataframne

```python
def get_wendu_type(x):
    if x['bWendu'] > 33:
        return 'high temperature'
    if x['yWendu'] < -10:
        return 'low temperature'
    return 'normal temperature'

data.loc[:, 'wendu_type'] = data.apply(get_wendu_type, axis = 1)
```

## .assign
Assign new columns to a DataFrame

Return a new object

```python
data.assign(
    yWendu_huashi = lambda x: x['yWendu'] * 9 /5 +32,
    bWendu_huashi = lambda x: x['bWendu'] * 9 /5 +32
)
```

## Assign in Condition

```python
data['wendu_type'] = ''
data.loc[data['wencha']>10, 'wencha_type'] = 'high Difference in Temperature'
data.loc[data['wencha']<10, 'wencha_type'] = 'normal Difference in Temperature'
```
