# map、apply、applymap

## Series.map

```python
import pandas as pd
stocks = pd.read_excel('datas/stocks/互联网公司股票.xlsx')
stocks.head()
```

```python
stocks['公司'].unique()
```

```python
dict_company_names = {
    "bidu": "百度",
    "baba": "阿里巴巴",
    "iq": "爱奇艺", 
    "jd": "京东"
}
```

#### Solution 1

```python
stocks['公司中文1'] = stocks['公司'].str.lower().map(dict_company_names)
```

```python
stocks.head()
```

#### Solution 2

```python
stocks['公司中文2'] = stocks['公司'].map(lambda x: dict_company_names[x.lower()] )
```

```python
stocks.head()
```

## Map on Transform between Series and DataFrame

```python
stocks["公司中文3"] = stocks["公司"].apply(
    lambda x : dict_company_names[x.lower()])
```

```python
stocks.head()
```

```python
stocks["公司中文4"] = stocks.apply(
    lambda x : dict_company_names[x["公司"].lower()], 
    axis=1)
```

```python
stocks.head()
```

## applymap

```python
sub_df = stocks[['收盘', '开盘', '高', '低', '交易量']]
```

```python
sub_df.head()
```

```python
sub_df.dtypes
```

```python
sub_df2 = sub_df.applymap(lambda x : int(x))
sub_df2
```

```python
sub_df2.dtypes
```

```python
stocks.loc[:, ['收盘', '开盘', '高', '低', '交易量']] = sub_df.applymap(lambda x : int(x))
```
