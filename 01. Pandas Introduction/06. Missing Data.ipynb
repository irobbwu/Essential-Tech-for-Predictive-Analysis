# Handling Missing Data

+ isnull and notnull: Detect wheather there is a null value
+ dropna: drop null value
    1. axis: delete rows or columns
    2. how: how = any(delete any null value); how = all(delete where all value are null)
    3. inplace: if False, it will create a new object
+ fillna: fill null value
    1. value: to replace by it
    2. method: ffill(using forward value to fii), bfill(using backward value to fill)
    3. axis: (0 or 'index')&(1 or 'columns)
    4. inplace: if False, it will create a new object

```python
import pandas as pd

studf = pd.read_excel('datas/student_excel/student_excel.xlsx', skiprows = 2)
studf
```

## isnull and notnull

```python
studf.isnull()
```

```python
studf['分数'].notnull()
```

## Dropnull

```python
studf.dropna(axis = 1, how = 'all', inplace = True)
```

```python
studf
```

```python
studf.dropna(axis =0, how = 'all', inplace = True)
```

```python
studf
```

## fillna

```python
studf.fillna({'分数':0})
studf
```

```python
studf.loc[:, '分数'] = studf['分数'].fillna(value = 0)
```

```python
studf
```

```python
studf.loc[:, '姓名'] = studf['姓名'].fillna(method = 'ffill')
studf
```

## Save

```python
studf.to_excel('datas/student_excel/student_excel_clean.xlsx',
              index = False)
```
